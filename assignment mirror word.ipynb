### Program to print the input word in reverse

#### Using a loop

```python
a=input("Enter a word: ")
i=1
print("mirrored word is: ",end="")
while i<=len(a):
    print(a[-i],end="")
    i+=1
```

#### without loop

```python
a=input("Enter a word: ")
print ("Your word is mirrored to:",a[::-1])
```

#### using reverse and join of a list

```python
a=input("Enter a word: ")
a1=list(a)
a1.reverse()
print("Your word is mirrored to:","".join(a1))
```
